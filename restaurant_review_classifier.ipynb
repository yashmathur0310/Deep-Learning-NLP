```python
from tensorflow import keras
```

```python
from keras.preprocessing.text import one_hot
from keras.preprocessing.sequence import pad_sequences
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import Embedding
from keras.layers import Flatten
```

```python
import pandas as pd 
```

```python
df=pd.read_csv('Restaurant_Reviews.tsv',sep='\t')
```

```python
df.head()
```

```python
import texthero as hero
def preprocessing(text):
    text=pd.Series(text)
    text=hero.clean(text)
    return text
```

```python
df['Review']=df['Review'].apply(lambda x:preprocessing(x))
```

```python
reviews=df['Review'].values
```

```python
vocab_size=250
```

```python
one_hot_reviews=[one_hot(d,vocab_size) for d in reviews]
```

```python
one_hot_reviews
```

```python
maxlength=15
```

```python
padded_reviews=pad_sequences(one_hot_reviews,maxlen=maxlength,padding='post')
```

```python
padded_reviews
```

```python
embeded_vocab_size=4
```

```python
model=Sequential([
    Embedding(vocab_size,embeded_vocab_size,input_length=maxlength),
    Flatten(),
    Dense(1,activation='sigmoid')
])
model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
x=padded_reviews
y=df['Liked'].values
```

```python
from sklearn.model_selection import train_test_split 
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=42,test_size=0.2)
```

```python
model.fit(x_train,y_train,epochs=100)
```

```python
model.predict(x)
```

```python
pred=model.predict(x_test)
```

```python
y_pred=[]
for elem in pred:
    if elem>=0.5:
        y_pred.append(1)
    else:
        y_pred.append(0)
```

```python
from sklearn.metrics import accuracy_score
ac=accuracy_score(y_test,y_pred)
ac
```
